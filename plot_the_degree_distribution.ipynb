```python
 #Compute and plot the degree distribution of a real-world network. Also compute local and global properties of the network.


import networkx as nx
import matplotlib.pyplot as plt
# BarabsiAlbert scale-free network
m=3
G= nx.barabasi_albert_graph(1000, m)
deg_freq = nx.degree_histogram(G)
print(deg_freq)
degree = range(len(deg_freq))
print (degree)
nx.draw(G, with_labels=True)
plt.figure(figsize=(19,10))
plt.loglog(degree[3:], deg_freq[3:], 'r*-')
plt. xlabel("degree")
plt. ylabel("frequency")
plt.title("Degree Distribution (Barabasi-Albert Model)")
plt. show()
# Global properties
print("\n--- Global Properties of Barabasi-Albert Graph ---")
print("Average degree:", sum(dict(G.degree()).values()) / G.number_of_nodes())
print("Average clustering coefficient:", nx.average_clustering(G))
print("Density:", nx.density(G))
print("Diameter:", nx.diameter(G))
print("Average shortest path length:", nx.average_shortest_path_length(G))
# WattsStrogatz small-world network
n=20
k=10
p= 0.5
G1 = nx.watts_strogatz_graph(n, k, p)
deg_freql = nx.degree_histogram(G1)
print(deg_freql)
degreel = range(len(deg_freql))
print (degreel)
nx.draw(G1, with_labels=True)
plt.figure(figsize=(10,10))
plt. loglog(degreel[3:], deg_freql[3:], 'r*-')
plt.xlabel( "degree")
plt.ylabel ("frequency")
plt.title("Degree Distribution (Watts-Strogatz Model)")
plt. show()
# Global properties
print("\n--- Global Properties of WattsStrogatz Graph ---")
print("Average degree:", sum(dict(G1.degree()).values()) / G1.number_of_nodes())
print("Average clustering coefficient:", nx.average_clustering(G1))
print("Density:", nx.density(G1))
print("Diameter:", nx.diameter(G1))
print("Average shortest path length:", nx.average_shortest_path_length(G1))
```
