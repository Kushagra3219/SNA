```python
# 1. Plot a weighted directed network such that the node size and edge width is proportional
# to the degree and edge weight respectively


import networkx as nx
import matplotlib.pyplot as plt
G = nx.DiGraph()
n = int(input("Enter number of edges: "))
print("Enter edges in the format: source target weight")
for i in range(n):
   u, v, w = input().split()
   G.add_edge(u, v, weight=int(w))
degrees = dict(G.degree())
node_sizes = [degrees[node] * 800 for node in G.nodes()]
weights = nx.get_edge_attributes(G, 'weight')
edge_weights = [weights[edge] for edge in G.edges()]
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=node_sizes, width=edge_weights, arrows=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=weights, font_color="red")
plt.title("Weighted Directed Network")
plt.show()
```
